```python
def print_bio(name, age, goal):
    print("*" * 25)
    print(f"Name: {name}, AGE: {age} Goal: {goal}")
    print("*" * 25)

print_bio("Martin", 20, "Complete This Course")
print_bio("Erling Haaland", 25, "Play Soccer" )
print_bio("Jay Gatsby", 32, "Win back Daisy")
```
