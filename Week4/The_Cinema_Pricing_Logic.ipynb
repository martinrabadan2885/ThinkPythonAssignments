```python
def calculate_ticket(age, hour):
    #If age <= 12 or age>= 65, set price to 8
    if age <= 12 or age >= 65:
        price = 8
    else:
        price = 12

    # If movie starts before 17, subtract 2 from price
    if hour < 17:
        price -= 2 # Subtract 2 from the calculated price

    return price

# Ask user for age
user_age = int(input("What is your age? "))

# Ask user for movie hour (24-hour format)
user_hour = int(input("What hour did you watch the movie? "))

# Call the calculate_ticket function with the obtained inputs
ticket_price = calculate_ticket(user_age, user_hour)

# Print the calculated ticket price
print("Your ticket price is $" + str(ticket_price))
```
